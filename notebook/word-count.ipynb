```python
from pyspark import SparkConf, SparkContext
sc = SparkContext(conf = SparkConf().setMaster('local').setAppName('WordCount'))
```

```python
book = sc.textFile('file:////Users/giovanna/Documents/GitHub/pyspark/SparkCourse/book.txt')
```

```python
import re
def normalizeWord (line):
    return re.compile(r'\W+', re.UNICODE).split(line.lower())
```

```python
words = book.flatMap(normalizeWord) # the returnin RDD do not have the same size, it could be larger or small
```

```python
count_words = words.countByValue()
for word, count in count_words.items():
    cleanWord = word.encode('ascii', 'ignore')
    if (cleanWord):
        print (str(cleanWord) + str(count))
```

```python
count_words = words.map(lambda x: (x, 1)).reduceByKey(lambda x,y: x+y)
count_words = count_words.map(lambda x: (x[1],x[0])).sortByKey(ascending=False)
results = count_words.collect()
```

```python
for count, word in results:
    cleanWord = word.encode('ascii', 'ignore')
    if (cleanWord):
        print (str(cleanWord) + str(count))
```
